```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return "[Count (step: " + str(self.step) + ")] " + str(self.count)
    
    def incr(self):
        self.count += self.step
    
    def __call__(self):
        self.incr()
        
    def __add__(self, other):
        self.count += other
        return Counter(self.count, self.step)
    
    def __sub__(self, other):
        self.count -= other
        return Counter(self.count, self.step)
    
    def __eq__(self, other):
        return self.count == other
    
    def __ne__(self, other):
        return self.count != other
    
    def __gt__(self,other):
        return self.count > other
    
    def __lt__(self,other):
        return self.count < other
        
a = Counter(10)
b = Counter(10,2)

print(a)
print(a>10)
print(a<10)
print(a==10)
print(a!=10)
```

요구사항 1은 __init__ 매직메소드를 이용해 생성자를 설정해서 구현한다.  
요구사항 2는 __str__ 매직메소드를 이용해 리턴값에 문자열을 넣어 구현한다.  
요구사항 3은 incr 이라는 메소드를 작성하고 self.count에 self.step만큼 더해서 구현한다.  
요구사항 4는 __call__ 매직메소드를 이용하고, 내부에서 self의 incr 메소드를 호출한다.  
요구사항 5는 __add__ __sub__ 매직메소드를 이용하고, 각 연삭을 하고 생성자를 반환해 구현한다.  
요구사항 6은 __eq__ __ne__ __gt__ __lt__ 매직메소드를 이용하고, 각 메소드는 연산을 리턴값으로 해 구현한다.

우선 elemicate_duplicate 메소드 부터 말하자면, @staticmethod라는 장식자가 붙은 정적메소드로, 인스턴스 객체 없이 MySet.elemicate_duplicate(I) 식으로도 호출할 수 있는 메소드다.  내용은 빈 리스트를 생성하고, I의 원소 중 중복내용이 없도록 추가하는 것이다. init 안에서 호출되어 self를 받는 경우 처리를 위해 I를 초기화하고, 중복내용이 없게 추가된 것을 I에 추가한다.  
__inti__ 의 경우는 받은 I의 내용을 하나씩 self의 리스트에 넣고 그 self를 elemicate_duplicate메소드의 인자로 넣어주는 내용의 생성자다.  
__str__ 은 result 라는 리스트를 일부를 초기화해서 선언하고, { 뒤로 self에 들어가있는 원소를 하나씩 추가하며, result를 갱신한다. 마지막의 ' ,' 를 지우기 위해 0부터 길이-2까지 뺀 만큼 슬라이싱하고 반환한다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __or__(self,other):
        l = self+other
        return MySet(l)
    
    def __and__(self,other):
        l = []
        for memi in self:
            if memi in other:
                l.append(memi)
        return MySet(l)
    
    def __sub__(self,other):
        l = self.copy()
        for memo in other:
            if memo in self:
                self.remove(memo)
        return MySet(l)

    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
```

요구사항 1은 __or__ 매직메소드를 이용한다. 새로운 리스트 l에 리스트 합연산을 하고 l을 생성자를 통해 반환하는 것으로 구현한다.  
요구사항 2는 __and__ 매직메소드를 이용한다. 새로운 리스트 l을 선언하고, 하나의 리스트의 요소를 뽑아 다른 리스트에 존재한다면, 그 요소를 l에 넣고 그 l을 생성자를 통해 반환하는 것으로 구현한다.  
요구사항 3은 __sub__ 매직메소드를 이용한다. 새로운 리스트 l에 copy함수를 활용해 self의 요소를 복사하고, other의 요소를 뽑아 없다면 그 요소를 지운다. l을 생성자를 통해 반환하는 것으로 구현한다.

```python
class SearchEngine(list):
    import requests
    import string
    import os
    
    def __init__(self, *url):
        for item in url:
            self.append(item)
            
    def addUrl(self, url):
        self.append(url)
    
    def removeUrl(self, url):
        self.remove(url)
    
    def listUrls(self):
        for item in self:
            print(item)
   
    # <script> 태그 제거
    def cleanhtml_script(self, htmlText):
        j = 0;
        S = ''
        for i in range(len(htmlText)):
            i = htmlText.find('<script',j)
            S += htmlText[j:i]
            j = htmlText.find('</script>',i)
            if(j == -1):
                break
            j += 9
        return S

    # <style> 태그 제거
    def cleanhtml_style(self, htmlText):
        j = 0;
        S = ''
        for i in range(len(htmlText)):
            i = htmlText.find('<style',j)
            S += htmlText[j:i]
            j = htmlText.find('</style>',i)
            if(j == -1):
                break
            j += 8         
        return S 

    def cleanhtml_class(self, htmlText):
        j = 0;
        S = ''
        for i in range(len(htmlText)):
            i = htmlText.find('class=',j)
            S += htmlText[j:i]
            j = htmlText.find('>',i)
            if(j == -1):
                break  
        return S 

    # 나머지 태그 제거
    def cleanhtml_ex(self, htmlText):
        j = 0;
        S = ''
        for i in range(len(htmlText)):
            i = htmlText.find('>',j)
            if(i == -1 or i == len(htmlText)-1):
                break
            j = htmlText.find('<',i)
            if htmlText[i+1] != '<':    
                S += htmlText[i+1:j] + " "
        return S

    # 태그제거 통합 & 리스트변환
    def cleanhtml(self, htmlText):
        source1 = self.cleanhtml_script(htmlText)
        source2 = self.cleanhtml_style(source1)
        source3 = self.cleanhtml_ex(source2)
        source4 = self.cleanhtml_class(source3)
        textList = source4.split()
        return textList

    # 구두문자 제거
    def cleanpunct(self, textList):
        for i, text in enumerate(textList):
            for p in self.string.punctuation:
                if p in text:
                    textList[i] = text.strip(self.string.punctuation)    

        return textList

    # 공백요소 제거
    def cleanblank(self, textList):
        cleanlist = ' '.join(textList).split()
        return cleanlist

    # 딕셔너리화
    def turndict(self, textList):
        textdict = {}
        for text in textList:
            if text not in textdict:
                textdict[text]=1
            else:
                textdict[text]+=1

        return textdict

    # 내용 : 불용어 추출.
    # 반환값 : 한글불용어, 영어불용어
    # 사용처 : erasestopword() 함수 내 사용
    def takestopword(self):
        file = open('stop_words_korean.txt','r', encoding = 'UTF8')
        LKoreanStopWord = []
        while 1:
            line = file.readline().strip()
            if not line: break
            splitline = line.split(' ')
            LKoreanStopWord.extend(splitline)

        file = open('stop_words_english.txt','r', encoding = 'UTF8')
        LEnglishStopWord = []
        while 1:
            line = file.readline().strip()
            if not line: break
            LEnglishStopWord.append(line)

        return LKoreanStopWord, LEnglishStopWord

    # 입력값 : 소스 내 단어 딕셔너리(A)
    # 내용 : A 내 불용어 삭제
    # 출력값 : 불용어가 제거된 A
    def erasestopword(self, DictSourceWord):
        LKoreanStopWord, LEnglishStopWord = self.takestopword()
        for dex in range(0, len(DictSourceWord)):    
            for stopword in LKoreanStopWord:
                if stopword in DictSourceWord:
                    del DictSourceWord[stopword]
            for stopword in LEnglishStopWord:
                if stopword in DictSourceWord:
                    del DictSourceWord[stopword]                                                
        return DictSourceWord

    # 입력값 : Url
    # 내용 : Url.html 파일 생성
    def makehtmlfile(self, Url):
        if not os.path.exists('Incremental'):
            os.mkdir('Incremental')
        i = Url.find('://')   
        SUrlname = './Incremental/' + Url[i+3:] + '.html'
        file = open(SUrlname,'w')
        file.close()

    # 입력값 : 불용어 제거된 딕셔너리
    # 출력값 : 가장 많이 나온 단어 리스트 [1번째, 2번째, 3번째]
    def picktopthree(DictSourceWord):
        Dsourceword = sorted(DictSourceWord.items(),reverse = True, key = lambda item: item[1])
        Ltopthree = [Dsourceword[0][0], Dsourceword[1][0], Dsourceword[2][0]]
        return Ltopthree

    # 출력값 : 찾을 단어 리스트
    def getsearchword():
        Sword = input('검색어를 입력하세요. : ')
        Lword = Sword.split()
        LKoreanStopWord, LEnglishStopWord = takestopword()    
        for stopword in LKoreanStopWord:
            if stopword in Lword:
                Lword.remove(stopword)
        for stopword in LEnglishStopWord:
            if stopword in Lword:
                Lword.remove(stopword)
        return Lword

    # 입력값 : 찾을 단어 리스트, url단어 리스트
    # 출력값 : url 단어리스트 속 찾는 단어:개수 와 총 개수가 입력된 딕셔너리
    def foundword(Lword, Durlword):
        Dword = {}
        for word in Lword:
            if word in Durlword:
                Dword[word] = Durlword.get(word)
        total = 0
        for count in Dword.values():
            total += count
        Dword['Total'] = total
        return Dword
            
    # 태그, 구두문자, 공백 제거. + 딕셔너리 변환
    def integratedmodule_eraseddict(self, source):     
        S1 = self.cleanhtml(source)
        S2 = self.cleanpunct(S1)
        S3 = self.cleanblank(S2)
        S4 = self.turndict(S3)
        S5 = self.erasestopword(S4)
        return S5
    
    def getWordsFrequency(self):
        if len(self) == 0:
            return {}
        diction = {}
        for i in self:
            req = self.requests.get(i)
            source = req.text
            dicttmp = self.integratedmodule_eraseddict(source)
            for i in dicttmp:
                if i in diction:
                    diction[i] += dicttmp[i] 
                else:
                    diction[i] = dicttmp[i]
        return diction
    
    def getMaxFreqencyWords(self):
        if len(self) == 0:
            return None
        diction = self.getWordsFrequency()
        sortwords = {}
        sortwords = sorted(diction.items(),reverse = True, key = lambda item: item[1])
        max = sortwords[0][1]
        topwords = {}
        for i in sortwords:
            if(i[1] != max):
                break
            topwords[i[0]] = i[1]
        print(topwords)
        
    def searchUrlByWord(self, word):
        max = 0
        syte = []
        for i in self:
            req = self.requests.get(i)
            source = req.text
            dicttmp = self.integratedmodule_eraseddict(source)
            if word in dicttmp:
                if dicttmp[word] > max:
                    syte = [i]
                if dicttmp[word] == max:
                    syte.append(i)
        for url in syte:
            print(url)
```
