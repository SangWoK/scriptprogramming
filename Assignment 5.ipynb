```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return "[Count (step: " + str(self.step) + ")] " + str(self.count)
    
    def incr(self):
        self.count += self.step
    
    def __call__(self):
        self.incr()
        
    def __add__(self, other):
        self.count += other
        return Counter(self.count, self.step)
    
    def __sub__(self, other):
        self.count -= other
        return Counter(self.count, self.step)
    
    def __eq__(self, other):
        return self.count == other
    
    def __ne__(self, other):
        return self.count != other
    
    def __gt__(self,other):
        return self.count > other
    
    def __lt__(self,other):
        return self.count < other
        
a = Counter(10)
b = Counter(10,2)

print(a)
print(a>10)
print(a<10)
print(a==10)
print(a!=10)
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __or__(self,other):
        l = self+other
        return MySet(l)
    
    def __and__(self,other):
        l = []
        for memi in self:
            if memi in other:
                l.append(memi)
        return MySet(l)
    
    def __sub__(self,other):
        l = self.copy()
        for memo in other:
            if memo in self:
                self.remove(memo)
        return MySet(l)

    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
```
